
## Turbilencia de Viento
### Escala longitud Turbulencia

La escala de la longitud de la turbulencia $L(z)$ representa el tamaño medio de ráfagas de viento. Para alturas por debajo de 200 m tenemos que:

- para $z>=z_{min}$   
$$ L(z)=L_t(\frac{z}{z_t})^\alpha$$

- para $z < z_{min}$  
$$ L(z)=L(z_{min})$$

con:
- para una altura de referencia $z_t=200m$ tenemos una escala de la longitud de la turbulencia de $L_t=300m$
- $\alpha=0,67+0,05ln(z_0)$

### Distribución del viento sobre frecuencias
La distribución en frecuencias del viento se expresa por la función adimensional de densidad espectral de potencias $S_L(z,n)$ que viene dada por:

$$S_L(z,n)=\frac{n S_V(z,n)}{\sigma_v^2}=\frac{6,8 f_L(z,n)}{(1+10,2f_L(z,n))^{5/3}}$$

siendo:
- $f_L(z,n)=\frac{n L(z)}{v_m(z)}$


```python
SL(f)=6.8*f/((1+10.2*f)^(5/3))
p1=plot(SL(f), (f,0.01, 10),color='red',axes_labels =['$f_L$','$S_L(f_L(z,n))$'],figsize=8,title='  Funcion desidad espectral de potencia ',scale="semilogx",gridlines=["minor",true],gridlinesstyle={'linestyle':'-','linewidth':0.2, 'color':'black'})
show(p1)
```


### Factor de Fondo

El factor de fondo (background factor) $B^2$ valora la falta de uniformidad en la acción de la presión sobre la estructura y se define como:
$$ B^2=\frac{1}{1+0.9(\frac{b+h}{L(z_e)})^{0.63}}$$
siendo:
- $b,h$ el ancho y alto de la estructura en cuestión
- $L(z_e)$ es la escala de la longitud de la turbulencia
